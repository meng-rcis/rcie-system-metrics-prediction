```python
import sys
sys.path.append('../')

import pandas as pd
import matplotlib.pyplot as plt

from constant.path import PATH
from constant.columns import COLS, INDEX_COL, FREQUENCY
```

```python
STARTING_INDEX = 180
ENDING_INDEX = 360

df = pd.read_csv(PATH, index_col=INDEX_COL, parse_dates=True, skiprows=1, header=None, names=COLS)
df = df[STARTING_INDEX:ENDING_INDEX]

skip_cols = ['timestamp', 'tps_error', 'status']

# Plotting each column
for column in df.columns:
    if column not in skip_cols:
        plt.figure()
        df[column].plot()
        # plt.title(column)
        plt.xlabel('Time')
        plt.ylabel(column)
        plt.show()
```
