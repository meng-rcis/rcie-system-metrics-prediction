```python
# Compare cutoff

first, second = 1, 2

df_1 = pd.read_csv(f'../dump/butterworth_observation/filtered_df_o3c{first}.csv')
df_2 = pd.read_csv(f'../dump/butterworth_observation/filtered_df_o3c{second}.csv')

# Plot all first 100 data with for loop
columns = ['cpu_usage', 'memory_usage', 'bandwidth_inbound', 'bandwidth_outbound', 'tps', 'tps_error', 'response_time']
for column in columns:
    plt.figure()
    plt.plot(original['time'][:shown], original[column][:shown], label='Original', color='lightgrey')
    plt.plot(df_1['time'][:shown], df_1[column][:shown], label=f'Norm Cutoff Freq: 0.{first}')
    plt.plot(df_2['time'][:shown], df_2[column][:shown], label=f'Norm Cutoff Freq: 0.{second}')
    plt.xlabel('Time')
    plt.ylabel(column)
    plt.title(column)
    plt.legend()
    plt.show()
```
