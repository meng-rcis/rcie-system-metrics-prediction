### Correlation

```python
import sys
sys.path.append('../')

# Load libraries
from utils.load import load_dataset
from constant.columns import FEATURES

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Load datasets
df = load_dataset()
df_features = df.drop(['time', 'timestamp', 'status'], axis=1)
```

```python
# Computing Point-Biserial Correlation
for col in FEATURES:
    print(f'Point-Biserial Correlation: {col}')
    correlation, p_value = stats.pointbiserialr(df.status, df[col])
    print(f'Point-Biserial Correlation: {correlation:.2f}')
    print(f'P-value: {p_value:.2f}')
    print('')
```

```python
# Plot correlation matrix
pearsoncorr = df_features.corr(method='pearson')
sns.heatmap(pearsoncorr,
xticklabels=pearsoncorr.columns,
yticklabels=pearsoncorr.columns,
cmap='RdBu_r',
annot=True,
linewidth=0.5)
plt.show()
```
